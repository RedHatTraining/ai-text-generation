```python
from transformers import pipeline, set_seed, AutoTokenizer
```

```python
tokenizer = AutoTokenizer.from_pretrained('.model')
generator = pipeline('text-generation', model='.model')
```

```python
generator(
        "OpenShift is",
        max_length=10,
        num_return_sequences=5,
        output_scores=True,
        return_full_text=False,
        do_sample=True,
        top_k=10,
        top_p=0.92
)
```
