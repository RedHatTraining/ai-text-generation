# Suggestions in AsciiDoc files with GTP-2

_GTP-2_ is an AI large language model capable of generating text, given an input string.

For example, given:

```
Hello
```

the model might return:

```
, World!
```

## What is GPT-2?

GPT-2, like its siblings (GPT, GPT-3, GPT-3.5), is a transformer model.
_Transformers_ are deep-learning models, which have proved to be particularly effective in generative machine learning tasks, such image or text generation.

* Transfomers are based on a concept called _self-attention_: the ability of a model to detect relationships between distant elements of the same input sequence.
